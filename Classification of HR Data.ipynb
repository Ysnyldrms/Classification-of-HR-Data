### Modelling

```python
x = pd.get_dummies(x,drop_first = True)   
```

```python
x
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
g = GaussianNB()
b = BernoulliNB()
```

```python
g.fit(x,y)                 # model trained 
```

```python
b.fit(x,y)                  
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
tahmin1 = g.predict(x)
tahmin2 = b.predict(x)           
```

```python
accuracy_score(tahmin1,y)
```

```python
accuracy_score(tahmin2,y)
```

```python
precision_score(tahmin2,y)
```

```python
confusion_matrix(tahmin2,y)                 # (True Positive, False Negative, False Positive, True Negative)
```

```python
confusion_matrix(tahmin1,y)
```

```python
print(classification_report(tahmin1,y))               # see as a report 
```

```python
print(classification_report(tahmin2,y))
```
